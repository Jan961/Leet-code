```python
import sqlite3
import random
import string
from pathlib import Path
import datetime
from datetime import timedelta
from PIL import Image
```

```python

#create fake puzzles

#helper functions
row_tuples= []

def create_random_str(length):
    alphabet = string.ascii_lowercase
    out_list= [0]*length
    for i in range(length):
        out_list[i] = alphabet[random.randint(0,25)]
    return "".join(out_list)


def convert_to_png(path: Path):
    im = Image.open(path)
    new_path = Path(str(path)[:-3] + "png")
    im.save(new_path)

    return new_path

def convertToBinaryData(filename):
    # Convert digital data to binary format
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData
```
