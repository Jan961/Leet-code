```python
import os
from pathlib import Path
```

```python
base_dir = Path(r'C:\Users\Owner\Desktop\bongard_raw')
authors = ["foundalis", "fairbanks", "bongard", "Hofstadter"]
```

```python

for author in authors:
    dir = base_dir.joinpath(author)

    for filename in os.listdir(dir):
        if str(filename)[-3:] == 'gif':
            os.remove(dir.joinpath(filename))
            print(f"{filename} removed \n")
```
